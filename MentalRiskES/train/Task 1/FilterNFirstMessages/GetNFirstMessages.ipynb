```python
with open(dir_path + "task1_gold_b.txt") as file:
  final_labels_list = file.readlines()
final_labels_list

final_labels = {}

for label in final_labels_list:
  if label == final_labels_list[0]:
    continue

  key = ""
  value = ""

  comma = False
  for letter in label:
    if letter == '\n':
      continue
    if letter == ',':
      comma = True
      continue
    if not comma:
      key += letter
    else:
      value += letter

  final_labels[key] = float(value)

final_labels
```

```python
import json
import os
import csv

number = 50

# Lista para almacenar los resultados
results = []

# Leer todos los archivos JSON en el directorio
for filename in os.listdir(dir_path + "train_task1/"):
    if filename.endswith(".json"):
        with open(os.path.join(dir_path + "train_task1/", filename), "r") as file:
            # Cargar el contenido del archivo JSON en una lista de diccionarios
            data = json.load(file)

            # Obtener el nombre del archivo sin la extensión ".json"
            name = os.path.splitext(filename)[0]
            print(len(data))

            if len(data) > number:
              # Concatenar los textos de los primeros number diccionarios
              text = ". ".join([d["message"] for d in data[:number]])
            else:
              text = ". ".join([d["message"] for d in data])

            # Añadir el resultado a la lista de resultados
            results.append((name, text, final_labels[name]))

# Escribir los resultados en un archivo CSV
with open(os.path.join(dir_path, f"output_{number}.csv"), "w", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(["Subject", "Text", "label"])
    writer.writerows(results)
```
