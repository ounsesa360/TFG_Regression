```python
import csv
from collections import Counter

# Función para leer el archivo CSV
def read_csv(filename):
    data = []
    with open(filename, 'r') as file:
        csv_reader = csv.reader(file)
        for row in csv_reader:
            data.append(row)
    return data

# Función para escribir en un archivo CSV
def write_csv(filename, data):
    with open(filename, 'w', newline='') as file:
        csv_writer = csv.writer(file)
        csv_writer.writerows(data)

# Nombre de los archivos de entrada y salida
input_filename = dir_path + 'task1_all_subjects_csv_with_labels.csv'
output_filename = dir_path + 'NumberMessagesWithLabel.csv'

# Leer el archivo CSV de entrada
input_data = read_csv(input_filename)

# Contar la cantidad de filas por Subject
subject_counter = Counter(row[0] for row in input_data)

# Generar los datos para el archivo de salida
output_data = [['Subject', 'Label', 'Number']]
input_data.remove(input_data[0])
subjects = {}
subjects_label = {}
for row in input_data:
    try:
      subjects[row[1]] = subjects[row[1]] + 1
    except:
      subjects[row[1]] = 1
    subjects_label[row[1]] = row[2]
print(subjects)
print(subjects_label)
for subject in subjects:
    output_data.append([subject, subjects_label[subject], subjects[subject]])
# Escribir en el archivo CSV de salida
write_csv(output_filename, output_data)

print("El archivo CSV de salida ha sido generado correctamente.")
```

```python
import csv
import matplotlib.pyplot as plt

# Función para leer el archivo CSV
def read_csv(filename):
    data = []
    with open(filename, 'r') as file:
        csv_reader = csv.reader(file)
        next(csv_reader)  # Omitir la primera fila si contiene encabezados
        for row in csv_reader:
            label = float(row[1])  # Convertir a float si es necesario
            number = float(row[2])  # Convertir a float si es necesario
            data.append((label, number))
    return data

# Nombre del archivo CSV
filename = dir_path + 'NumberMessagesWithLabel.csv'

# Leer el archivo CSV
data = read_csv(filename)

# Extraer los valores de x y y
x = [point[0] for point in data]
y = [point[1] for point in data]

# Crear el gráfico de dispersión
plt.scatter(x, y)
plt.xlabel('Label')
plt.ylabel('Número de mensajes')

# Mostrar el gráfico
plt.show()
```
